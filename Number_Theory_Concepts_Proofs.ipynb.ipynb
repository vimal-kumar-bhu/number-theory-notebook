## 1. GCD and Euclidean Algorithm

The greatest common divisor of two integers a and b is the largest integer dividing both.

Euclidean Algorithm:
gcd(a, b) = gcd(b, a mod b)

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

print(gcd(48, 18))
```

### Proof

Let d = gcd(a, b).

Then d divides both a and b.

Since:
a = bq + r

Any common divisor of a and b must also divide r.

Thus:
gcd(a, b) = gcd(b, r)

Hence proved.

## 2. Modular Arithmetic

We say:
a ≡ b (mod n) if n divides (a − b)

Used to simplify large computations.

```python
print(pow(2, 10, 7))
```

### Proof Idea

2^10 = 1024

1024 mod 7 = 2

Also:
2^3 ≡ 1 (mod 7)

So:
2^9 ≡ 1  
2^10 ≡ 2 (mod 7)

## 3. Euler’s Totient Function

φ(n) counts integers ≤ n that are coprime with n.

```python
def phi(n):
    result = n
    p = 2
    while p*p <= n:
        if n % p == 0:
            while n % p == 0:
                n //= p
            result -= result // p
        p += 1
    if n > 1:
        result -= result // n
    return result

print(phi(10))
```

### Proof (for prime p)

Numbers less than p:
1,2,...,p−1

Since p has no divisors except 1 and itself,
all are coprime.

Thus:
φ(p) = p − 1

## 4. Chinese Remainder Theorem

Solve:
x ≡ a1 (mod n1)
x ≡ a2 (mod n2)

If gcd(n1, n2)=1, solution exists.

```python
def crt(a1, n1, a2, n2):
    for x in range(n1*n2):
        if x % n1 == a1 and x % n2 == a2:
            return x

print(crt(2,3,3,5))
```

### Proof Idea

Since gcd(n1,n2)=1:

m1*n1 + m2*n2 = 1

Construct:
x = a1*m2*n2 + a2*m1*n1

This satisfies both congruences.

## 5. Prime Numbers

A prime number has only two divisors: 1 and itself.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

print(is_prime(17))
```

### Proof Idea

If n is composite, it has a factor ≤ √n.

Thus checking up to √n is sufficient.

## 6. Linear Diophantine Equation

ax + by = c

Solution exists iff gcd(a,b) divides c.

```python
import math

a, b, c = 2, 3, 7
print(math.gcd(a, b))
```

### Proof

Let d = gcd(a, b)

Then:
ax + by = d

If d divides c, multiply equation to get solution.

Hence solution exists iff d | c.
